```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Financial Statement Analysist
# - Income Statement (revenues, profitst)
# - Balance Sheet 
# - Cash Flow

# Income Statement
income_statement = pd.read_csv("income-statement.csv", index_col='Date')

income_statement["Profit"].plot(label="Profit")
income_statement["Revenue"].plot(label="Revenue")
plt.ylabel("Cash (Billions)")
plt.title("INCOME STATEMENT")
plt.legend()
plt.show()
```

```python
balance_sheet = pd.read_csv("balance-sheet.csv", index_col='Date')

current_assets = balance_sheet['Current Assets']
current_liabilities = balance_sheet['Current Liabilities']
net_asset = balance_sheet['Net Assets']
goodwill  = balance_sheet['Goodwill']

balance_sheet
```

```python
# Balance Sheet
current_assets.plot(label="current assets")
current_liabilities.plot(label="current liabilities")
plt.legend()
plt.show()
```

```python
net_asset.plot(label="Net Asset")
goodwill.plot(label="Goodwill")
plt.title("Net Asset and Goodwill")
plt.legend()
plt.show()
```

```python
# Cash Flow
cash_flow = pd.read_csv("cash-flow-statement.csv")
cash_flow["Net Cash Flow"].plot()
plt.title("Net Cash Flow")
plt.show()
```
