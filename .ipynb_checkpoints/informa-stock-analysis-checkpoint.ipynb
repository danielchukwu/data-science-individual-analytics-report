<h3>Training and Test sets</h3>

```python
# Training sets
INFTr = informa.iloc[0:473]
EBTr  = eventbrite.iloc[0:473]
```

```python
# Test sets
INFTs = informa.iloc[473:504]
EBTs  = eventbrite.iloc[473:504]
```

<h3>Daily Returns</h3>

```python
# INFTr & EBTr - Training sets
returns1 = (INFTr['Adj Close'] - INFTr['Adj Close'].shift(1)) / INFTr['Adj Close'].shift(1)
returns2 = (EBTr['Adj Close']  - EBTr['Adj Close'].shift(1))  / EBTr['Adj Close'].shift(1)
```

```python
# add return column to Training sets
INFTr['Return'] = returns1
EBTr['Return']  = returns2
```

```python
# INFTs & EBTs - Test sets
returns1 = (INFTs['Adj Close'] - INFTs['Adj Close'].shift(1)) / INFTs['Adj Close'].shift(1)
returns2 = (EBTs['Adj Close']  - EBTs['Adj Close'].shift(1))  / EBTs['Adj Close'].shift(1)
```

```python
# add return column to Test sets
INFTs['Return'] = returns1
EBTs['Return']  = returns2
```

```python
# print current Training and Test sets
print(INFTr.head())
print(EBTr.head())
print(INFTs.head())
print(EBTs.head())
```

```python
# Drop all rows containing NaN from Training and Test sets 
# INFTr.dropna(inplace=True)
# INFTs.dropna(inplace=True)
# EBTr.dropna(inplace=True)
# EBTs.dropna(inplace=True)

# Option 2: fill all Nan rows with the average value from their column (also known as series)
INFTr.fillna(INFTr['Return'].mean(), inplace=True)
INFTs.fillna(INFTs['Return'].mean(), inplace=True)
EBTr.fillna(EBTr['Return'].mean(), inplace=True)
EBTs.fillna(EBTr['Return'].mean(), inplace=True)
```

```python
# Drop all rows containing NaN from Training and Test sets
INFTr.drop(columns=["Open", "High", "Low", "Close", "Volume"], inplace=True)
INFTs.drop(columns=["Open", "High", "Low", "Close", "Volume"], inplace=True)
EBTr.drop(columns=["Open", "High", "Low", "Close", "Volume"], inplace=True)
EBTs.drop(columns=["Open", "High", "Low", "Close", "Volume"], inplace=True)
```

<h3>Regression</h3>

```python
# # The two dataframes have the same column names, we should change the names and merge them together
INFTr.columns = ['INF_Price','INF_Return']
EBTr.columns = ['EB_Price','EB_Return']

#concatenate them by columns
OverallTraining=pd.concat([INFTr, EBTr],axis=1)
```

```python
OverallTraining.head()
```

```python
OverallTraining.plot.scatter(x='INF_Price', y='EB_Price', alpha=0.7)
plt.show()
```

```python
pd.plotting.scatter_matrix(OverallTraining.iloc[:,[0,2]], alpha=0.7)
plt.show()
```

```python
OverallTraining.iloc[:,[0,2]].corr(method='pearson')
```

```python
pd.plotting.scatter_matrix(OverallTraining.iloc[:,[1,3]], alpha=0.7)
plt.show()
```

```python
OverallTraining.iloc[:,[1,3]].corr(method='pearson')
```

```python
INF_EB_reg = sfa.ols('INF_Price ~ EB_Price', data=OverallTraining).fit()
```

```python
INF_EB_reg.summary()
```

```python
sa.graphics.plot_fit(INF_EB_reg, exog_idx='EB_Price', vlines=False)
plt.ylabel("INF Price")
plt.xlabel("EB Price")
plt.title("INF vs EB")
plt.show()
```

<h3>Time Series</h3>

```python
INFTs.columns = ['INF_Price','INF_Return']
EBTs.columns = ['EB_Price','EB_Return']
```

<h3>Moving Average</h3>

```python
ma5 = INFTr.rolling(20).mean()
```

```python
ma20 = INFTr.rolling(50).mean()
```

```python
INFTr['INF_Price'].plot(figsize=(15,10), title= 'Moving average', label='Price')
ma5['INF_Price'].plot(label='MA5')
ma20['INF_Price'].plot(label='MA20')
plt.legend()
plt.ylabel('Price',fontsize=15)
plt.show()
```

```python
# create forecasts
# INFma5fcsts = pd.DataFrame({'INF_Price': [ma5['INF_Price'][-1]]*20, 'Date': INFTr.index[0:20]})
INFma5fcsts = pd.DataFrame({'INF_Price': [ma5['INF_Price'][-1]]*20 }, index=INFTs.index[0:20])
```

```python
INFTr['INF_Price'].plot(figsize=(15,10),label='train',title='5-day Moving Average')
INFTs['INF_Price'].plot(label='test')
ma5['INF_Price'].plot(label='fitted')
INFma5fcsts.plot(label='forecasts')
plt.legend()
plt.show()
```

```python
MAE(INFTs['INF_Price'],INFma5fcsts[0:20])
```

```python
MSE(INFTs['INF_Price'],INFma5fcsts[0:20])
```

```python
print('MAE: {:,.4f}'.format(MAE(INFTs['INF_Price'],INFma5fcsts[0:20])), '\nMSE: {:,.4f}'.format(MSE(INFTs['INF_Price'],INFma5fcsts[0:20])))
```

### Single Exponential Smoothing

```python
INFses=SES(INFTr['INF_Price'],initialization_method="estimated").fit()
```

```python
INFsesfcsts=INFses.forecast(20)
INFsesfcsts.index=INFTs.index[0:20]
```

```python
INFTr['INF_Price'].plot(figsize=(15,10),marker='o',color='black',label='train',title='Single Exponential Smoothing')
INFTs['INF_Price'].plot(marker='o',color='brown',label='test')
INFses.fittedvalues.plot(marker='v', color='blue',label=r'$\alpha=%s$'%INFses.model.params['smoothing_level'])
INFsesfcsts.plot(marker='v', color='green',label='forecasts')
plt.legend()
plt.show()
```

```python
print(
    'MAE: {:,.4f}'.format(MAE(INFTs['INF_Price'],INFsesfcsts)),
    '\nMSE: {:,.4f}'.format(MSE(INFTs['INF_Price'],INFsesfcsts))
)
```

### Double Exponential Smoothing

```python
INFdes=DES(INFTr['INF_Price'],initialization_method="estimated").fit()
```

```python
INFdesfcsts=INFdes.forecast(20)
INFdesfcsts.index=INFTs.index
```

```python
INFTr['INF_Price'].plot(figsize=(15,10),marker='o',color='black',label='train',title='Double Exponential Smoothing')
INFTs['INF_Price'].plot(marker='o',color='brown',label='test')
INFdes.fittedvalues.plot(marker='v', color='blue',label='fitted')
INFdesfcsts.plot(marker='v', color='green',label='forecasts')
plt.legend()
plt.show()
```

```python
print(
    'MAE: {:,.4f}'.format(MAE(INFTs['INF_Price'],INFdesfcsts)),
    '\nMSE: {:,.4f}'.format(MSE(INFTs['INF_Price'],INFdesfcsts))
)
```

### Holt's Winters (Additive)

```python
INFtesa=TES(INFTr['INF_Price'],seasonal_periods=4, trend='add', seasonal='add', use_boxcox=True, initialization_method="estimated").fit()
INFtesfcstsa=INFtesa.forecast(20)
INFtesfcstsa.index=INFTs.index
```

```python
INFTr['INF_Price'].plot(figsize=(15,10),marker='o',color='black',label='train',title="Holt's Winters Method (Additive)")
INFTs['INF_Price'].plot(marker='o',color='brown',label='test')
INFtesa.fittedvalues.plot(marker='v', color='blue',label='fitted')
INFtesfcstsa.plot(marker='v', color='green',label='forecasts')
plt.legend()
plt.show()
```

```python
print(
    'MAE: {:,.4f}'.format(MAE(INFTs['INF_Price'],INFtesfcstsa)),
    '\nMSE: {:,.4f}'.format(MSE(INFTs['INF_Price'],INFtesfcstsa))
)
```

### Holt's Winters (Multiplicative)

```python
INFtesm=TES(INFTr['INF_Price'],seasonal_periods=4, trend='add', seasonal='mul', use_boxcox=True, initialization_method="estimated").fit()
INFtesfcstsm=INFtesm.forecast(20)
INFtesfcstsm.index=INFTs.index
```

```python
INFTr['INF_Price'].plot(figsize=(15,10),marker='o',color='black',label='train',title="Holt's Winters Method (Multiplicative)")
INFTs['INF_Price'].plot(marker='o',color='brown',label='test')
INFtesm.fittedvalues.plot(marker='v', color='blue',label='fitted')
INFtesfcstsm.plot(marker='v', color='green',label='forecasts')
plt.legend()
plt.show()
```

```python
print(
    'MAE: {:,.4f}'.format(MAE(INFTs['INF_Price'],INFtesfcstsm)),
    '\nMSE: {:,.4f}'.format(MSE(INFTs['INF_Price'],INFtesfcstsm))
)
```
